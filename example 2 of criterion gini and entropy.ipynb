```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\malware\data\salaries.csv')
df
```

```python
#let us convet the columns into numeric with labelEncoder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#create new column
df['company_n']=le.fit_transform(df['company'])
df['job_n']=le.fit_transform(df['job'])
df['degree_n']=le.fit_transform(df['degree'])
df
```

```python
df=df.drop(['company','job','degree'],axis='columns')
df
```

```python
#divide the data into x and y
x=df.drop(['salary_more_than_100k'],axis='columns')
x
```

```python
y=df[['salary_more_than_100k']]
y
```

```python
#create the Decision Tress Classifier model
#default critierion is 'gini'.We can use criterion='entropy' also
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
model.predict([[1,1,0]])
```

```python
#create the Decision Tress Classifier model
# criterion='entropy' also
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model.fit(x,y)
```

```python
model.predict([[1,1,0]])
```
